## 일원배치법/일원분산분석 (One-Way ANOVA)

- 한 개의 요인(factor)과 해당 요인의 여러 수준(level) 간의 그룹 평균이 서로 유의미하게 다른지 검정하는 통계 기법

[예시]
- 서로 다른 교육 방법(강의식, 토론식, 실습식)이 학생들의 시험 점수에 미치는 영향을 비교하고 싶을 때.
- 세 가지 브랜드의 세탁기가 세제를 사용하는 효율이 다른지 확인하고 싶을 때.

[귀무가설, 대립가설]
- 귀무가설 : 모든 그룹의 평균이 동일하다.
- 대립가설 : 적어도 하나의 그룹 평균이 다르다.


```python
from scipy.stats import f_oneway

# 데이터 준비
group1 = [23, 20, 22, 21, 24]
group2 = [30, 28, 29, 31, 27]
group3 = [40, 38, 39, 37, 41]

# 일원분산분석 수행
f_oneway(group1, group2, group3)
```

```python
from scipy.stats import f_oneway

# 데이터 준비
group1 = [23, 20, 22, 21, 24]
group2 = [30, 28, 29, 31, 27]
group3 = [40, 38, 39, 37, 41]

# 일원분산분석 수행
stat, p_value = f_oneway(group1, group2, group3)

# 결과 출력
print(f"F-statistic: {stat}, p-value: {p_value}")

# 해석
if p_value < 0.05:
    print("귀무가설 기각: 그룹 간 평균이 유의하게 다릅니다.")
else:
    print("귀무가설 채택: 그룹 간 평균이 유의하게 다르지 않습니다.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 준비
data = {
    "score": group1 + group2 + group3,
    "group": ["Group1"] * len(group1) + ["Group2"] * len(group2) + ["Group3"] * len(group3),
}
df = pd.DataFrame(data)

# 박스 플롯
sns.boxplot(x="group", y="score", data=df)
plt.title("Group-wise Score Distribution")
plt.show()
```

## 이원분산분석 (Two-Way ANOVA)

[검정 대상]
1. 각 요인별 주효과(Main Effect)
- 요인 1: 교육 방법이 점수에 미치는 주효과
- 요인 2: 성별이 점수에 미치는 주효과

2. 상호작용 효과(Interaction Effect)
- 교육 방법과 성별의 상호작용이 점수에 미치는 효과

[귀무가설, 대립가설]
- 요인 1의 귀무가설: 요인 1의 각 수준 간 평균 차이가 없다.
- 요인 2의 귀무가설: 요인 2의 각 수준 간 평균 차이가 없다.
- 상호작용의 귀무가설: 두 요인 간 상호작용 효과가 없다.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 예제 데이터 생성
data = {
    "score": [85, 90, 88, 92, 78, 82, 80, 76, 75, 79, 77, 73],
    "method": ["A", "A", "A", "A", "B", "B", "B", "B", "C", "C", "C", "C"],
    "gender": ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F", "M", "F"],
}

df = pd.DataFrame(data)

# 이원분산분석 모델 생성
model = ols("score ~ C(method) + C(gender) + C(method):C(gender)", data=df).fit() # 범주형 데이터 : C(칼럼) 형태
anova_table = sm.stats.anova_lm(model, typ=2)

# 결과 출력
print(anova_table)
```

- p-value < 0.05: 교육 방법은 점수에 유의미한 영향을 미칩니다.
- p-value < 0.05: 성별은 점수에 유의미한 영향을 미칩니다.
- p-value > 0.05: 교육 방법과 성별 간의 상호작용은 유의하지 않습니다.

## t검정 단일표본

ttest_1samp(집단, 모평균, alternative = 'greater' / 'less' / 'two_sided'(기본))

- 목적: 표본 평균이 특정 기준값(μ₀)과 통계적으로 유의미하게 다른지를 검정.
- 귀무가설 : 표본의 평균이 기준값과 같다는 가설.
- 대립가설(양측 검정) : 표본 평균이 기준값과 다르다.
- 대립가설(단측 검정) : 표본 평균이 기준값보다 크다. or 표본 평균이 기준값보다 작다.

```python
from scipy.stats import ttest_1samp

# 샘플 데이터
sample = [18, 20, 22, 21, 19, 20, 23]

# 모평균
pop_mean = 20

# t-검정 수행
stat, p_value = ttest_1samp(sample, pop_mean, alternative='two-sided')

# 결과 출력
print(f"T-statistic: {stat}, p-value: {p_value}")

# 해석
if p_value < 0.05:
    print("귀무가설 기각: 샘플 평균은 모평균과 유의미하게 다릅니다.")
else:
    print("귀무가설 채택: 샘플 평균은 모평균과 유의미하게 다르지 않습니다.")
```

## 점추정, 구간추정

- 점추정 = 표본 평균
- 구간추정 : 

```python
import numpy as np
import scipy.stats as stats

# 표본 데이터
data = [50, 55, 45, 60, 65, 70, 40, 55, 50, 60]

# 점추정 (표본 평균)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # 자유도 n-1 사용
n = len(data)

# 구간추정 (95% 신뢰구간)
confidence_level = 0.95
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)
margin_of_error = t_critical * (sample_std / np.sqrt(n))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# 결과 출력
print(f"점추정 (표본 평균): {sample_mean:.2f}")
print(f"구간추정 (95% 신뢰구간): ({lower_bound:.2f}, {upper_bound:.2f})")
```

## 베르누이 분포, 이항 분포

[베르누이 분포 (Bernoulli Distribution)]
- 베르누이 분포 : 성공(1) 또는 실패(0) 두 가지 결과만 가능한 단일 실험의 확률 분포.

[예시]
- 동전을 한 번 던져 앞면이 나올 확률.
- 특정 질병에 대해 치료가 성공할 확률.

[이항 분포 (Binomial Distribution)]
- 베르누이 시행을 n번 반복하여 성공한 횟수의 확률 분포.

[예시]
- 동전을 10번 던져 앞면이 나오는 횟수.
- 시험 문제 20개 중 맞춘 문제 수.

```python
import numpy as np
from scipy.stats import bernoulli, binom

# 데이터: 각 시행의 성공(1) 또는 실패(0) 기록
data = [1, 0, 1, 1, 0, 1, 0, 1, 0, 1]  # 총 10번 시행

# 1. 베르누이 분포
# 성공 확률 (p) 추정
p_hat = np.mean(data)  # 성공한 횟수 / 전체 시행 횟수
print(f"[베르누이 분포] 성공 확률 (p): {p_hat:.2f}")

# 확률 질량 함수 계산
x_bernoulli = [0, 1]  # 가능한 값 (0: 실패, 1: 성공)
pmf_bernoulli = bernoulli.pmf(x_bernoulli, p_hat)
print(f"[베르누이 분포] PMF: {dict(zip(x_bernoulli, pmf_bernoulli))}")

# 2. 이항 분포
# 전체 시행 횟수와 성공 횟수 추출
n = len(data)  # 전체 시행 횟수
k = sum(data)  # 성공 횟수

# 이항 분포의 확률 계산
x_binom = np.arange(0, n + 1)  # 가능한 성공 횟수 (0부터 n까지)
pmf_binom = binom.pmf(x_binom, n, p_hat)
print(f"\n[이항 분포] PMF (성공 횟수):")
for x, p in zip(x_binom, pmf_binom):
    print(f"  성공 횟수 {x}: {p:.4f}")
```

### 1-3. 만약 age 변수가 5단위 증가하면 오즈비(Odds ratio)는 몇 배로 변화하는가?

```python
from statsmodels.api import Logit
import numpy as np

# 로지스틱 회귀 모델
formula = 'Survived ~ Pclass + Age + Parch + SibSp'
results = Logit.from_formula(formula, df).fit()

# Age가 5단위 증가할 때의 오즈비 계산
age_coef = results.params['Age']  # Age의 회귀 계수
odds_ratio = np.exp(5 * age_coef)  # 5단위 증가 시 오즈비 변화

print(f"Age가 5단위 증가하면 오즈비는 약 {odds_ratio:.4f}배로 변화합니다.")
```

### 2-3) 1)에서 만든 모델에 x1 = 1, x2 = 30, x3 = 2, x4 = 1, x5 = 50 값을 대입하면 Y값은 얼마로 예측되겠는가? -> 정답 : 0.763

```python
import pandas as pd
import statsmodels.api as sm


# 결측값 제거
df = df[['Survived', 'Pclass', 'Age', 'Parch', 'SibSp', 'Fare']].dropna()

# 변수 이름 소문자로 변경
df.columns = [col.lower() for col in df.columns]

# OLS 모델 적합
formula = "survived ~ pclass + age + parch + sibsp + fare"
model = sm.OLS.from_formula(formula, df).fit()

# 회귀계수 확인
print(model.summary())

# 새로운 값 대입
new_data = pd.DataFrame({
    'pclass': [1],  # X1
    'age': [30],    # X2
    'parch': [2],   # X3
    'sibsp': [1],   # X4
    'fare': [50]    # X5
})

# 예측값 계산
y_pred = model.predict(new_data)
print(f"예측된 Y값: {y_pred.iloc[0]:.4f}")
```
