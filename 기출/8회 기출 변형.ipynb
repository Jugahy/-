### 1-1. 주어진 데이터 기준으로 와인 소비량이 가장 많은 대륙(Continent) 내에서 와인 소비량이 2번째로 많은 국가(Country)의 와인 소비량을 구하시오.

```python
import pandas as pd
import numpy as np

data = [
    ["SouthKorea", 40.2, "Asia"],
    ["Japan", 50.5, "Asia"],
    ["China", 25.3, "Asia"],
    ["India", 10.4, "Asia"],
    ["Thailand", 35.6, "Asia"],
    ["Germany", 110.7, "Europe"],
    ["France", 90.8, "Europe"],
    ["Italy", 60.1, "Europe"],
    ["Spain", 70.4, "Europe"],
    ["Belgium", 85.2, "Europe"],
    ["USA", 95.7, "NorthAmerica"],
    ["Canada", 80.5, "NorthAmerica"],
    ["Mexico", 50.3, "NorthAmerica"],
    ["Brazil", 65.5, "SouthAmerica"],
    ["Argentina", 45.7, "SouthAmerica"],
    ["Chile", 30.8, "SouthAmerica"],
    ["SouthAfrica", 55.6, "Africa"],
    ["Nigeria", 20.5, "Africa"],
    ["Egypt", 15.2, "Africa"],
    ["Australia", 105.8, "Oceania"],
    ["NewZealand", 94.5, "Oceania"]
]

df = pd.DataFrame(data, columns=["Country", "와인소비량", "Continent"])

df.head()
```

```python
# 와인 소비량이 가장 많은 대륙 찾기 위해 대륙 별 와인소비량 합 구하기

df2 = df.groupby("Continent").sum()
df2
```

```python
# 대륙별 와인소비량이 가장 큰 값을 갖는 대륙명 찾기

continent_ = df2[df2["와인소비량"] == max(df2["와인소비량"])].index[0]
continent_
```

```python
# 찾은 Europe에 속하는 Country 출력하기

df3 = df[df["Continent"] == continent_[0]]
df3
```

```python
# Europe에 속하는 나라의 와인소비량에 따른 내림차순 정렬

sorted_df = df3.sort_values("와인소비량", ascending=False)
sorted_df
```

### 1-2. 다음은 각 국가의 목적별 방문객 수를 나타낸 데이터이다.총 방문객 수 대비 관광목적의 여행자 비율이 3번째로 높은 국가의 사업 목적의 방문자수를 A, 관광 목적의 여행객이 2번째로 많은 국가의 공무 목적 방문자수를 B라고 할 때 A + B 값은 무엇인가?

```python
import pandas as pd

data = [
    ["SouthKorea", 800, 150, 200],
    ["Japan", 900, 100, 220],
    ["China", 700, 180, 250],
    ["India", 600, 120, 210],
    ["Thailand", 750, 130, 180],
    ["Germany", 845, 110, 240],
    ["France", 780, 140, 200],
    ["Italy", 820, 170, 220],
    ["Spain", 740, 160, 210],
    ["Belgium", 790, 150, 190],
    ["USA", 950, 140, 240],
    ["Canada", 860, 130, 230],
    ["Mexico", 710, 120, 220],
    ["Brazil", 780, 110, 210],
    ["Australia", 900, 100, 210]
]

df = pd.DataFrame(data, columns=["국가", "관광", "공무", "사업"])

# 데이터 확인
df.head()
```

```python
# 총 방문객 수 개비 관광목적의 여행자 비율을 알기 위해 total, ratio 컬럼 생성

df["total"] = df["관광"] + df["공무"] + df["사업"]
df["ratio"] = df["관광"] / df["total"]

df
```

```python
# 관광목적 여행자의 비율이 3번째로 높은 나라를 찾고 그 나라를 사업을 목적으로 방문한 사람의 수 뽑기

A = df.sort_values("ratio", ascending=False)
A = A.iloc[2]["사업"]
A
```

```python
# 관광 목적의 여행객이 2번째로 많은 국가의 공무 목적 방문자수 찾기

B = df.sort_values("관광", ascending=False)
B = B.iloc[1]["공무"]
B
```

```python
A + B
```

### 1-3. 주어진 데이터에서 관광변수와 공무변수를 각각 Min-max scaling을 진행한 다음 공무변수의 표준편차 - 관광변수의 표준편차 값을 구하시오. (단, 정답은 반올림하여 소수점 첫째자리까지 구하시오.)
* MinMaxScaler
* 표준편차

```python
import pandas as pd

data = [
    ["SouthKorea", 800, 150, 200],
    ["Japan", 900, 100, 220],
    ["China", 700, 180, 250],
    ["India", 600, 120, 210],
    ["Thailand", 750, 130, 180],
    ["Germany", 845, 110, 240],
    ["France", 780, 140, 200],
    ["Italy", 820, 170, 220],
    ["Spain", 740, 160, 210],
    ["Belgium", 790, 150, 190],
    ["USA", 950, 140, 240],
    ["Canada", 860, 130, 230],
    ["Mexico", 710, 120, 220],
    ["Brazil", 780, 110, 210],
    ["Australia", 900, 100, 210]
]

df = pd.DataFrame(data, columns=["국가", "관광", "공무", "사업"])

# 데이터 확인
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 진행
scaler = MinMaxScaler()

df[["관광", "공무"]] = scaler.fit_transform(df[["관광", "공무"]])

# 표준편차 계산
std_a = df["공무"].std()
std_b = df["관광"].std()

# 결과 계산
result = std_a - std_b
print(round(result, 1))
```

### 제 2유형 기출 문제파악
* 문제: 지하철 탑승인원수를 예측하기
* 회귀 or 분류: 회귀 문제
* 성능지표: MAE(Mean Absolute Error)


### 기출 데이터셋 주요 정보
* 독립변수(X): 11개(날씨, 날짜 등)
* 종속변수(Y): 지하철 탑승인원수(정수형)
* 데이터셋: train 약 12,000개 / test 약 2,000개
* 결측치: 없음


### 특이사항
* name 변수(지하철역명) 삭제시 감점처리 되는 경우, 30점 감점(10/40점)
* id, 특정 명칭 컬럼 함부로 삭제하면 안 됨.


### 데이터셋 설명

* temperature: 기온 (5도 ~ 35도 사이)
* humidity: 습도 (20% ~ 90% 사이)
* precipitation: 강수 여부 (0: 없음, 1: 있음)
* wind_speed: 풍속 (0 ~ 10 m/s)
* day_of_week: 요일 (0: 일요일 ~ 6: 토요일)
* is_holiday: 공휴일 여부 (0: 평일, 1: 공휴일)
* is_weekend: 주말 여부 (0: 평일, 1: 주말)
* hour: 시간대 (0시 ~ 23시)
* promotion: 승차장 내 프로모션 여부 (0: 없음, 1: 있음)
* school_vacation: 방학 여부 (0: 아님, 1: 방학)
* special_event: 특별 행사 여부 (0: 없음, 1: 있음)
* passenger_count: 지하철 탑승 인원
* name: 지하철 역 이름

```python
# 데이터 생성 (수정금지)
import numpy as np
import pandas as pd

# 랜덤 시드 설정
np.random.seed(42)

# 데이터 수
n = 1000

# 독립변수 생성 (날씨, 날짜 등)
data = {
    'temperature': np.random.uniform(5, 35, n),  # 기온 (5도 ~ 35도 사이)
    'humidity': np.random.uniform(20, 90, n),   # 습도 (20% ~ 90% 사이)
    'precipitation': np.random.choice([0, 1], n, p=[0.7, 0.3]),  # 강수 여부
    'wind_speed': np.random.uniform(0, 10, n),  # 풍속 (0 ~ 10 m/s)
    'day_of_week': np.random.choice(range(7), n),  # 요일 (0: 월요일 ~ 6: 일요일)
    'is_holiday': np.random.choice([0, 1], n, p=[0.8, 0.2]),  # 공휴일 여부
    'is_weekend': np.random.choice([0, 1], n, p=[0.7, 0.3]),  # 주말 여부
    'hour': np.random.choice(range(24), n),  # 시간대 (0시 ~ 23시)
    'school_vacation': np.random.choice([0, 1], n, p=[0.8, 0.2]),  # 방학 여부
    'special_event': np.random.choice([0, 1], n, p=[0.95, 0.05])  # 특별 행사 여부
}

# 종속변수로 탑승 인원수 생성 (임의 공식 사용)
data['passenger_count'] = (
    300
    + 5 * data['temperature']
    - 2 * data['humidity']
    + 50 * data['precipitation']
    - 10 * data['wind_speed']
    - 20 * data['is_holiday']
    - 15 * data['is_weekend']
    + 3 * data['hour']
    + 50 * data['school_vacation']
    + 80 * data['special_event']
    + np.random.normal(0, 30, n)  # 노이즈 추가
)

# 종속변수는 정수형으로 변환
data['passenger_count'] = np.round(data['passenger_count']).astype(int)

# 데이터프레임으로 변환
df = pd.DataFrame(data)

# 독립변수 중에서 소수점이 있는 변수는 소수점 첫째 자리까지 반올림
df['temperature'] = np.round(df['temperature'], 1)
df['humidity'] = np.round(df['humidity'], 1)
df['wind_speed'] = np.round(df['wind_speed'], 1)

# 지하철 2호선 역 이름 추가
station_names = [
    '시청', '을지로입구', '을지로3가', '을지로4가', '동대문역사문화공원', '신당',
    '성수', '건대입구', '구의', '강변', '잠실나루', '잠실', '잠실새내', '종합운동장',
    '교대', '서초', '방배', '사당', '낙성대', '서울대입구', '봉천', '신림', '신대방',
    '문래', '영등포구청', '당산', '합정', '홍대입구', '신촌', '이대', '아현'
]

# 1000개 중 2호선의 역이 43개이므로 순환하며 채우기
name_variable = [station_names[i % len(station_names)] for i in range(1000)]

# 'name' 변수 추가
df['name'] = name_variable

# 범주형 데이터를 카테고리형으로 변환
categorical_columns = ['precipitation', 'day_of_week', 'is_holiday', 'is_weekend', 'school_vacation', 'special_event']
df[categorical_columns] = df[categorical_columns].astype('category')

# Train, Test 셋 나누기
train = df.iloc[:700]  # 초기 700개를 train으로
test = df.iloc[700:]  # 마지막 300개를 test로
y_test = test['passenger_count']  # 실제 시험에서는 주어지지 않습니다.
test = test.drop(['passenger_count'], axis=1)

# Train 데이터 확인
train.head(3)
```

```python
import pandas as pd
import numpy as np

import seaborn as sns
df = sns.load_dataset("titanic")

df = df[["survived", "pclass", "age", "parch", "sibsp", "fare"]]
df = df.dropna()

df.head()
```

### 3-1. 로지스틱 회귀분석 (15점, 각 소문제당 5점)
**<span style="color: #2D3748; background-color:#fff5b1;"> 주요 키워드: 유의한 변수, 회귀계수, 오즈비</span>**

survived를 종속변수(Y), 'pclass', 'age', 'parch', 'sibsp', 'fare' 변수를 독립변수(X)로 분석했을 때 아래 질문에 답하시오.
(단, 모든 답은 반올림하여 소수점 둘째자리까지 구하시오)

* 1-1. 전체 데이터를 모두 활용하여 로지스틱회귀분석 진행 후에 유의하지 않은 변수의 수는?

* 1-2. 1번 문제에서 유의한 변수만 사용하여 로지스틱회귀분석을 진행했을 때 회귀계수의 합계를 구하시오.

* 1-3. 만약 age 변수가 5단위 증가하면 오즈비(Odds ratio)는 몇 배로 변화하는가?

```python
# 1-1) 0.05보다 작으면 유의하고 0.05보다 크면 유의하지 않은 변수이다. -> 정답 : 1개

from statsmodels.api import Logit

formula = "survived ~ pclass + age + parch + sibsp + fare"
results = Logit.from_formula(formula, df).fit()

print(results.summary())
```

```python
# 1-2) coef : 회귀계수 / coef에 있는 모든 값 그냥 더하기 -> 정답 : 2.45

from statsmodels.api import Logit

formula = "survived ~ pclass + age+ parch + sibsp"
results = Logit.from_formula(formula, df).fit()

print(results.summary())
print(round(3.7520-1.2663-0.0452+0.2808-0.2733, 2))
```

```python
# 1-3) A단위 증가하면 오즈비는 몇 배 증가하냐고 물어보면 A를 coef에 곱해서 exp 해주면 됨.
# 정답 : 0.8

print(round(np.exp(5*-0.0452), 2))
```

### 3-2. 다중선형 회귀분석 (15점, 각 소문제당 5점)

**<span style="color: #2D3748; background-color:#fff5b1;"> 주요 키워드: p-value, 회귀계수, Rsq, 예측값 구하기</span>**
* 2-1) 모델링 후 p-value가 가장 큰 변수의 회귀계수 값을 구하시오. -> 정답 : 0.0006

* 2-2) 1)에서 구한 다중선형회귀 모델의 결정계수(Rsq)를 구하시오. -> 정답 : 0.195

* 2-3) 1)에서 만든 모델에 x1 = 1, x2 = 30, x3 = 2, x4 = 1, x5 = 50 값을 대입하면 Y값은 얼마로 예측되겠는가? </br>-> 정답 : 0.763

```python
# 2-1) p-value가 가장 큰 fare의 coef가 정답
# 2-2) 결정계수 = R-squared 이니 0.195가 정답

from statsmodels.api import OLS

formula = "survived ~ pclass + age + parch + sibsp + fare"
results = OLS.from_formula(formula, df).fit()

print(results.summary())
```

```python
# 2-3) 다중선형회귀 식에 주어진 값 대입하면 끝

def function(x1, x2, x3, x4, x5):
    
    y = 1.1936 + (-0.2432*x1) + (-0.0086*x2) + (0.0492*x3) + (-0.0575*x4) + (0.0006*x5)
    
    return y

print(function(1, 30, 2, 1, 50))
```
