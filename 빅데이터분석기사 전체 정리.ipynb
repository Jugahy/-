(1분기:1,2,3월 / 2분기 : 4,5,6월 / 3분기 7,8,9월 / 4분기 10,11,12월 , 1분기 월평균 : 1,2,3월의 총범죄수 평균)

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e6_p1_3.csv')
df.head(5)
```

```python
df["총범죄수"] = df["절도"] + df["사기"] + df["배임"] + df["방화"] + df["폭행"]

ls = []

for k in ["2018", "2019"]:
    for i in range(1,5):
        for j in range(3):
            if i == 1:
                ls.append(k+"_1")
            elif i == 2:
                ls.append(k+"_2")
            elif i == 3:
                ls.append(k+"_3")
            elif i == 4:
                ls.append(k+"_4")
                
df["분기"] = ls

df.groupby("분기")["총범죄수"].mean().sort_values()

df[df["분기"] == "2019_2"].sort_values("사기", ascending=False).iloc[0]["사기"]
```

```python
# 출력을 원하실 경우 print() 함수 활용
# 예시) print(df.head())

# getcwd(), chdir() 등 작업 폴더 설정 불필요
# 파일 경로 상 내부 드라이브 경로(C: 등) 접근 불가

import pandas as pd

train = pd.read_csv("data/customer_train.csv")
test = pd.read_csv("data/customer_test.csv")

# 사용자 코딩

# 1. 데이터 분리
x_train = train.drop(["회원ID", "성별"], axis=1)
y = train["성별"]

x_test = test.drop(["회원ID"], axis=1)


# 2. 결측치 처리
x_train["환불금액"] = x_train["환불금액"].fillna(0)
x_test["환불금액"] = x_test["환불금액"].fillna(0)

# 3. 수치형 변수 스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num = x_train.select_dtypes(exclude="object").columns
x_train[num] = scaler.fit_transform(x_train[num])
x_test[num] = scaler.transform(x_test[num])

# 4. 범주형 변수 스케일링
from sklearn.preprocessing import LabelEncoder

scaler = LabelEncoder()

x_train["주구매상품"] = scaler.fit_transform(x_train["주구매상품"])
x_test["주구매상품"] = scaler.transform(x_test["주구매상품"])
x_train["주구매지점"] = scaler.fit_transform(x_train["주구매지점"])
x_test["주구매지점"] = scaler.transform(x_test["주구매지점"])

# 5. train, test 데이터 나누기
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y, test_size=0.2)

# 모델 학습
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train, y_train)
y_val_pred = model.predict(x_val)



# 결과 확인
from sklearn.metrics import roc_auc_score
auc_score = roc_auc_score(y_val, y_val_pred)
print(auc_score)

# 예측 결과 저장
y_pred = model.predict(x_test)
result = pd.DataFrame(y_pred, columns=["pred"])
result.to_csv("result.csv", index=False)

# 결과 확인
result = pd.read_csv("result.csv")
print(result)







```

# 8회 2유형

```python
# 데이터 생성 (수정금지)
import numpy as np
import pandas as pd

# 랜덤 시드 설정
np.random.seed(42)

# 데이터 수
n = 1000

# 독립변수 생성 (날씨, 날짜 등)
data = {
    'temperature': np.random.uniform(5, 35, n),  # 기온 (5도 ~ 35도 사이)
    'humidity': np.random.uniform(20, 90, n),   # 습도 (20% ~ 90% 사이)
    'precipitation': np.random.choice([0, 1], n, p=[0.7, 0.3]),  # 강수 여부
    'wind_speed': np.random.uniform(0, 10, n),  # 풍속 (0 ~ 10 m/s)
    'day_of_week': np.random.choice(range(7), n),  # 요일 (0: 월요일 ~ 6: 일요일)
    'is_holiday': np.random.choice([0, 1], n, p=[0.8, 0.2]),  # 공휴일 여부
    'is_weekend': np.random.choice([0, 1], n, p=[0.7, 0.3]),  # 주말 여부
    'hour': np.random.choice(range(24), n),  # 시간대 (0시 ~ 23시)
    'school_vacation': np.random.choice([0, 1], n, p=[0.8, 0.2]),  # 방학 여부
    'special_event': np.random.choice([0, 1], n, p=[0.95, 0.05])  # 특별 행사 여부
}

# 종속변수로 탑승 인원수 생성 (임의 공식 사용)
data['passenger_count'] = (
    300
    + 5 * data['temperature']
    - 2 * data['humidity']
    + 50 * data['precipitation']
    - 10 * data['wind_speed']
    - 20 * data['is_holiday']
    - 15 * data['is_weekend']
    + 3 * data['hour']
    + 50 * data['school_vacation']
    + 80 * data['special_event']
    + np.random.normal(0, 30, n)  # 노이즈 추가
)

# 종속변수는 정수형으로 변환
data['passenger_count'] = np.round(data['passenger_count']).astype(int)

# 데이터프레임으로 변환
df = pd.DataFrame(data)

# 독립변수 중에서 소수점이 있는 변수는 소수점 첫째 자리까지 반올림
df['temperature'] = np.round(df['temperature'], 1)
df['humidity'] = np.round(df['humidity'], 1)
df['wind_speed'] = np.round(df['wind_speed'], 1)

# 지하철 2호선 역 이름 추가
station_names = [
    '시청', '을지로입구', '을지로3가', '을지로4가', '동대문역사문화공원', '신당',
    '성수', '건대입구', '구의', '강변', '잠실나루', '잠실', '잠실새내', '종합운동장',
    '교대', '서초', '방배', '사당', '낙성대', '서울대입구', '봉천', '신림', '신대방',
    '문래', '영등포구청', '당산', '합정', '홍대입구', '신촌', '이대', '아현'
]

# 1000개 중 2호선의 역이 43개이므로 순환하며 채우기
name_variable = [station_names[i % len(station_names)] for i in range(1000)]

# 'name' 변수 추가
df['name'] = name_variable

# 범주형 데이터를 카테고리형으로 변환
categorical_columns = ['precipitation', 'day_of_week', 'is_holiday', 'is_weekend', 'school_vacation', 'special_event']
df[categorical_columns] = df[categorical_columns].astype('category')

# Train, Test 셋 나누기
train = df.iloc[:700]  # 초기 700개를 train으로
test = df.iloc[700:]  # 마지막 300개를 test로
y_test = test['passenger_count']  # 실제 시험에서는 주어지지 않습니다.
test = test.drop(['passenger_count'], axis=1)

# Train 데이터 확인
train
```

```python
# 1. 데이터 나누기
x_train = train.drop(["passenger_count", "name"], axis=1)
y = train["passenger_count"]

x_test = test.drop(["name"], axis=1)


# 2. 결측치 처리
x_train.isnull().sum()
x_test.isnull().sum()
x_train["wind_speed"] = x_train["wind_speed"].fillna(x_train["wind_speed"].mean())


# 3. 수치형 변수 스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num = x_train.select_dtypes(exclude="category").columns
x_train[num] = scaler.fit_transform(x_train[num])
x_test[num] = scaler.transform(x_test[num])


# 4. 범주형 변수 스케일링 (이 데이터는 라벨인코딩 다 되어있는 상태)


# 5. train, test 데이터 나누기
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y, test_size=0.2)


# 6. 모델 학습
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train, y_train)
y_val_pred = model.predict(x_val)


# 7. 평가
from sklearn.metrics import mean_absolute_error

mse = mean_absolute_error(y_val, y_val_pred)


# 8. 예측 결과 저장
y_pred = model.predict(x_test)
result = pd.DataFrame(y_pred, columns=["pred"])
result.to_csv("result.csv", index=False)


# 9. 결과 확인
result = pd.read_csv("result.csv")
print(result)
```

```python
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv")

print(train.shape)
display(train.head())
test.head()
```

```python
# 1. 데이터 나누기
x_train = train.drop(["ID", "이용금액"], axis=1)
y = train["이용금액"]

x_test = test.drop(["ID"], axis=1)


# 2. 결측치 처리 (없음.)


# 3. 수치형 변수 스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num = x_train.select_dtypes(exclude="object").columns
x_train[num] = scaler.fit_transform(x_train[num])
x_test[num] = scaler.transform(x_test[num])

# 4. 범주형 변수 스케일링
from sklearn.preprocessing import LabelEncoder

scaler = LabelEncoder()

x_train["업종명"] = scaler.fit_transform(x_train["업종명"])
x_test["업종명"] = scaler.transform(x_test["업종명"])
x_train["성별"] = scaler.fit_transform(x_train["성별"])
x_test["성별"] = scaler.transform(x_test["성별"])

x_train = pd.get_dummies(x_train, columns=["이용자구분"])
x_test = pd.get_dummies(x_test, columns=["이용자구분"])

# 5. train, test 나누기
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y, test_size=0.2)


# 6. 모델 학습
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train, y_train)
y_val_pred = model.predict(x_val)

# 7. 평가
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))


# 8. 예측한 값 저장
y_pred = model.predict(x_test)
results = pd.DataFrame(y_pred, columns=["pred"])
results.to_csv("results.csv", index=False)


# 9. 저장된 파일 확인
result = pd.read_csv("results.csv")
result
```

# 6회 2유형

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/ep6_p2_test.csv')

display(train.head())
test.head()
```

```python
# 1. 데이터 나누기
x_train = train.drop(["ID", "General_Health"], axis=1)
y = train["General_Health"]

x_test = test.drop(["ID"], axis=1)


# 2. 결측치 처리 (없음.)


# 3. 수치형 변수 스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num = x_train.select_dtypes(exclude="object").columns


x_train[num] = scaler.fit_transform(x_train[num])
x_test[num] = scaler.transform(x_test[num])


# 4. 범주형 변수 스케일링 (LabelEncoder : Checkup, Exercise, Heart_Disease, Skin_Cancer, Other_Cancer, Depression, Diabetes, Arthritis, Sex, Age_Category, Smoking_History)
from sklearn.preprocessing import LabelEncoder

scaler = LabelEncoder()

object_ = ["Checkup", "Exercise", "Heart_Disease", "Skin_Cancer", "Other_Cancer", "Depression", "Diabetes", "Arthritis", "Sex", "Age_Category", "Smoking_History"]

for i in object_:
    x_train[i] = scaler.fit_transform(x_train[i])
    x_test[i] = scaler.transform(x_test[i])


# 5. train, test 나누기
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y, test_size=0.2)


# 6. 모델 학습
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train, y_train)
y_val_pred = model.predict(x_val)

# 7. 평가
from sklearn.metrics import f1_score

f1 = f1_score(y_val, y_val_pred ,average="macro")
print(f1)

# 8. 예측값 저장
y_pred = model.predict(x_test)
results = pd.DataFrame(y_pred, columns=["pred"])
results.to_csv("results.csv", index=False)

# 9. 예측값 확인
results = pd.read_csv("results.csv")
results
```

### 3-2. 주어진 데이터를 이용하여 다중선형회귀 문제를 해결하시오.

```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/data_6_3_2.csv')

df.head()
```

### 1) 선형회귀 모델을 만들어 독립변수 Ozone의 회귀계수를 구하시오. -> 정답 : 0.172

### 2) Solar와 Ozone은 고정, Wind가 증가하면 따라 Temp가 감소하는지 검증하기 위한 다중선형회귀 분석을 수행하여 Wind의 회귀계수에 대한 p-value를 구하시오. (유의수준 0.05) -> 정답 : 0.169

```python
from statsmodels.api import OLS

formula = "Temp ~ Solar + Wind + Ozone"
results = OLS.from_formula(formula, df).fit()

print(results.summary())
```

### 3) Solar가 150, Wind가 7.5, Ozone이 45일 경우, 예측 값과 그에 대한 95% 신뢰구간을 구하시오. -> 정답 : 예측값은 78.8263, 95% 신뢰구간 하한 77.112, 상한 80.540

```python
data = pd.DataFrame({"Solar":[150], "Wind":[7.5], "Ozone":[45]})

prediction = results.get_prediction(data)
print(prediction.summary_frame())
```
